```python
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize
import base64
from io import BytesIO
import requests

target = []
images = []
flat_data = []

DATADIR ='/content/images'
CATEGORIES =['sunflowers','cricketball','ice cream cone']

for category in CATEGORIES:
  class_num = CATEGORIES.index(category)
  path = os.path.join(DATADIR,category)
  for img in os.listdir(path):
    img_array=imread(os.path.join(path,img))
    #print(img_array.shape)
    #plt.imshow(img_array)
    img_resized=resize(img_array,(150,150,3))
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(class_num)

flat_data=np.array(flat_data)
target=np.array(target)
images=(np.array(images))
```

```python
flat_data[0]
```

```python
unique,count=np.unique(target,return_counts=True)
plt.bar(CATEGORIES,count)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(flat_data,target,
                                               test_size=0.3,random_state=109)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid = [
              {'C':[1,10,100,1000],'kernel':['linear']},
              {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']},
]

svc = svm.SVC(probability=True)
clf = GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```python
y_pred =clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
import pickle
pickle.dump(clf,open('img_model.p','wb'))
```

```python
model =pickle.load(open('img_model.p','rb'))
```

```python
flat_data = []
url = input('Enter your image URL (base64 format): ')

# Remove the "data:image/jpeg;base64," part
if url.startswith("data:image"):
    base64_str = url.split(",")[1]  # Extract base64 part
    img_data = base64.b64decode(base64_str)
    img = imread(BytesIO(img_data))  # Read image from byte stream
else:
    # If it's a normal URL
    response = requests.get(url)
    img = imread(BytesIO(response.content))

# Resize and prepare image data
img_resized = resize(img, (150, 150, 3))
flat_data.append(img_resized.flatten())
flat_data = np.array(flat_data)

# Show original and resized image shapes
print("Original image shape:", img.shape)
print("Resized image shape:", img_resized.shape)

# Display the resized image
plt.imshow(img_resized)
plt.axis('off')
plt.show()

# Make a prediction
y_out = model.predict(flat_data)
y_out = CATEGORIES[y_out[0]]
print(f'PREDICTED OUTPUT: {y_out}')
```
